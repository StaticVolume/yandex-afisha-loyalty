Ищем и удаляем дубликаты

````python
def find_dups():   
    full_dups = df.duplicated(keep=False).sum()
    display(f"Полные дубликаты: {full_dups} ({full_dups/len(df)*100:.2f}%)")
    
    partial_dups = df.duplicated(subset=['user_id', 'order_id', 'order_ts'], keep=False).sum()
    display(f"Частичные дубликаты (user_id + order_id + order_ts): {partial_dups}")
    
    if 'order_id' in df.columns:
        order_dups = df[df.duplicated(subset=['order_id'], keep=False)]
        display(f"Заказы с повторяющимися order_id: {len(order_dups)}")
    if not order_dups.empty:  # безопасная проверка
        display(order_dups[['order_id', 'user_id', 'order_ts', 'revenue_rub']].head(10))
    if len(order_dups) > 0:
        display(order_dups[['order_id', 'user_id', 'order_ts', 'revenue_rub']].head(10))
        
    return (full_dups, partial_dups)
````

````python
def clean_dubs(df):
    df = df.drop_duplicates(keep='first')
    full_dups = df.duplicated(keep=False).sum()
    display(f"Полных дубликатов : {full_dups}")
````

````python
full_dups, partial_dups = find_dups()
if full_dups == 0 and partial_dups == 0:
    display("Дубликаты не обнаружены — данные чистые, можно переходить к анализу")
else:
    display(f"Найдены дубликаты — требуется дополнительная очистка")
    clean_dubs(df)    
````

 - Отфильтруем значения ≤ 0 (возвраты не относятся к целевой метрике продаж).
 - Обрежем верхний хвост по 99-му перцентилю.

````python
def plot_distribution_with_outliers(df, column, threshold=None, title_suffix='', 
                                    xlim=None, bins=100, log_scale=True):

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    #ГИСТОГРАММА (левый график)
    axes[0].hist(df[column], 
                 bins=bins, 
                 color='green', 
                 edgecolor='black', 
                 alpha=0.7, 
                 log=log_scale)
    
    # Добавляем линию перцентиля, если передана
    if threshold is not None:
        axes[0].axvline(threshold, 
                       color='red', 
                       linestyle='--', 
                       linewidth=2, 
                       label=f'99% Перцентиль: {threshold:.0f} руб.')
        axes[0].legend()
    
    axes[0].set_xlabel('Выручка, RUB')
    axes[0].set_ylabel('Количество заказов (Log)' if log_scale else 'Количество заказов')
    axes[0].set_title(f'Распределение {column} {title_suffix}')
    axes[0].grid(axis='y', alpha=0.3)
    
    # BOXPLOT (правый график)
    axes[1].boxplot(df[column].dropna(), 
                   vert=False, 
                   patch_artist=True,
                   boxprops=dict(facecolor='green', color='black'),
                   whiskerprops=dict(color='black'),
                   capprops=dict(color='black'),
                   medianprops=dict(color='red', linewidth=2))
    
    axes[1].set_xlabel('Выручка, RUB')
    axes[1].set_title(f'Диаграмма размаха (Boxplot) {column} {title_suffix}')
    
    # Ограничиваем ось X, если передано
    if xlim is not None:
        axes[1].set_xlim(xlim)
    
    axes[1].grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
````

````python
plot_distribution_with_outliers(
    df, 
    'revenue_rub', 
    threshold=threshold,
    title_suffix='(до фильтрации)',
    xlim=(-100, 5000), 
    bins=100,
    log_scale=True
)
````

Признаки, описывающие первый заказ пользователя, помогают:
- Куда инвестировать бюджет
- Улучшить онбординг новых пользователей
- Выявить «узкие места» в воронке привлечения
- Разделить пользователей для персонализированных коммуникаций

Что будем анализировать:
- first_event_type — жанр первого мероприятия 
- first_device — тип устройства
- first_region — география
- first_partner — билетный оператор

````python
def analyze_segment(df, column, title, top_n=None):
    
    # Подсчёт пользователей по категориям
    segment_stats = df[column].value_counts().reset_index()
    segment_stats.columns = [column, 'user_count']
    segment_stats['share_pct'] = (segment_stats['user_count'] / len(df) * 100).round(2)
    
    # Сортировка по убыванию
    segment_stats = segment_stats.sort_values('user_count', ascending=False)
    
    # Ограничение топ-N для наглядности (если категорий много)
    if top_n:
        plot_data = segment_stats.head(top_n)
    else:
        plot_data = segment_stats
    
    # Визуализация
    plt.figure(figsize=(10, 5))
    bars = plt.barh(plot_data[column].astype(str), plot_data['user_count'], color='steelblue', edgecolor='black')
    
    # Добавление процентов на график
    for i, (count, pct) in enumerate(zip(plot_data['user_count'], plot_data['share_pct'])):
        plt.text(count + max(plot_data['user_count'])*0.01, i, f'{pct:.1f}%', va='center')
    
    plt.xlabel('Количество пользователей')
    plt.ylabel(title)
    plt.title(f'Распределение пользователей по {title}')
    plt.gca().invert_yaxis() 
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return segment_stats
````

- Ответьте на вопросы:
    - Какие сегменты пользователей чаще возвращаются на Яндекс Афишу?
    - Наблюдаются ли успешные «точки входа» — такие сегменты, в которых пользователи чаще совершают повторный заказ, чем в среднем по выборке?

При интерпретации результатов учитывайте размер сегментов: если в сегменте мало пользователей (например, десятки), то доли могут быть нестабильными и недостоверными, то есть показывать широкую вариацию значений.

---


````python
def analyze_retention_by_segment(df, column, title, top_n=None):
    """
    Анализирует долю возвращающихся пользователей (2+ заказа) по сегментам.
    """
    # Группировка по сегменту
    retention_stats = df.groupby(column).agg(
        user_count=('user_id', 'count'),
        retention_rate=('is_two', 'mean') 
    ).reset_index()
    
    retention_stats['retention_rate_pct'] = (retention_stats['retention_rate'] * 100).round(2)
    retention_stats = retention_stats.sort_values('user_count', ascending=True)
    
    # Ограничение топ-N для визуализации
    if top_n:
        plot_data = retention_stats.head(top_n)
    else:
        plot_data = retention_stats
    
    # Средняя доля возврата по всей выборке
    avg_retention = df['is_two'].mean() * 100
    
    # Визуализация
    plt.figure(figsize=(12, 6))
    
    # Цвета: зелёный если выше среднего, красный если ниже
    colors = ['green' if r > avg_retention else 'red' for r in plot_data['retention_rate_pct']]
    
    bars = plt.barh(range(len(plot_data)), plot_data['retention_rate_pct'], color=colors, edgecolor='black', alpha=0.8)
    
    # Добавление значений на график
    for i, (rate, count) in enumerate(zip(plot_data['retention_rate_pct'], plot_data['user_count'])):
        plt.text(rate + 1, i, f'{rate:.1f}%\n(n={count})', va='center', fontsize=9)
    
    # Линия среднего
    plt.axvline(avg_retention, color='blue', linestyle='--', linewidth=2, label=f'Среднее: {avg_retention:.1f}%')
    
    plt.yticks(range(len(plot_data)), plot_data[column].astype(str))
    plt.xlabel('Доля пользователей с 2+ заказами, %')
    plt.ylabel(title)
    plt.title(f'Возврат пользователей по {title}\n(зелёный = выше среднего, красный = ниже среднего)')
    plt.gca().invert_yaxis()
    plt.legend()
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return retention_stats, avg_retention
````

````python
day_map = {
    0: 'понедельник', 1: 'вторник', 2: 'среда', 3: 'четверг',
    4: 'пятница', 5: 'суббота', 6: 'воскресенье'
}

# Извлекаем номер дня и сразу переводим в русское название
user_profile['first_day_of_week'] = user_profile['first_order_date'].dt.dayofweek.map(day_map)

# Считаем статистику
day_stats = user_profile.groupby('first_day_of_week').agg(
    users=('user_id', 'count'),
    return_rate=('is_two', 'mean')
).reset_index()

day_stats['return_rate_pct'] = (day_stats['return_rate'] * 100).round(1)

# 3. Сортируем
days_order = ['понедельник', 'вторник', 'среда', 'четверг', 'пятница', 'суббота', 'воскресенье']
day_stats['first_day_of_week'] = pd.Categorical(day_stats['first_day_of_week'], categories=days_order, ordered=True)
day_stats = day_stats.sort_values('first_day_of_week')

print("Статистика по дням недели первого заказа:")
display(day_stats)

# Визуализация
plt.figure(figsize=(10, 5))

# Столбцы с цветом: зелёный если выше среднего, красный если ниже
colors = ['green' if r > 61.67 else 'red' for r in day_stats['return_rate_pct']]
bars = plt.bar(day_stats['first_day_of_week'], day_stats['return_rate_pct'], 
               color=colors, edgecolor='black', alpha=0.8)

# Подписи значений
for bar in bars:
    h = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, h + 1, f'{h:.1f}%', ha='center', fontsize=9)

# Линия среднего
plt.axhline(61.67, color='blue', linestyle='--', linewidth=2, label=f'Среднее: 61.67%')

plt.xlabel('День недели первого заказа')
plt.ylabel('Доля возврата, %')
plt.title('Влияние дня недели первого заказа на возврат пользователей')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

best = day_stats.loc[day_stats['return_rate_pct'].idxmax()]
worst = day_stats.loc[day_stats['return_rate_pct'].idxmin()]
diff = day_stats['return_rate_pct'].max() - day_stats['return_rate_pct'].min()

display(f"Вывод:")
display(f"Лучший день: {best['first_day_of_week']} ({best['return_rate_pct']:.1f}%)")
display(f"Худший день: {worst['first_day_of_week']} ({worst['return_rate_pct']:.1f}%)")
display(f"Размах колебаний: {diff:.1f} п.п.")

if diff < 5:
    display("Разница небольшая (< 5 п.п.) — день недели слабо влияет на возврат")
else:
    display(f"Разница заметная ({diff:.1f} п.п.) — день недели может влиять на лояльность")
````

День недели первого заказа слабо влияет на возврат: разница между лучшим (суббота, 64.2%) и худшим (четверг, 59.6%) днём составляет всего 4.6 п.п., что незначительно для бизнес-решений — пользователи возвращаются стабильно независимо от дня первой покупки

Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;

- Визуализируйте результат корреляции с помощью тепловой карты.

````python
# Выбираем признаки для анализа
features = [
    'total_orders',          
    'avg_revenue_rub',        
    'avg_tickets',            
    'avg_days_between',       
    'first_device',          
    'first_event_type',       
    'first_region',           
    'first_partner',          
    'is_two',                 
    'is_five'                
]

# Копируем данные и убираем пропуски в числовых колонках
df_corr = user_profile[features].copy()
df_corr = df_corr.dropna(subset=['avg_revenue_rub', 'avg_tickets', 'avg_days_between'])

#Указываем числовые (интервальные) колонки
interval_cols = ['avg_revenue_rub', 'avg_tickets', 'avg_days_between', 'total_orders']

corr_matrix = phik_matrix(df_corr, interval_cols=interval_cols)

#Визуализация: тепловая карта
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, linewidths=0.5, cbar_kws={'label': 'phi_k'})
plt.title('Корреляция признаков с количеством заказов (phi_k)')
plt.tight_layout()
plt.show()

#Топ-признаки по связи с total_orders
display(f"Признаки, связанные с количеством заказов:")
target_corr = corr_matrix['total_orders'].drop('total_orders').sort_values(ascending=False)
display(target_corr)
````

````python
#Сегментация по количеству заказов
def order_segment(x):
    if x == 1:
        return '1 заказ'
    elif 2 <= x <= 4:
        return '2-4 заказа'
    else:
        return '5+ заказов'

df_corr['order_segment'] = df_corr['total_orders'].apply(order_segment)

#Повторяем анализ для каждого сегмента
for segment in ['1 заказ', '2-4 заказа', '5+ заказов']:
    subset = df_corr[df_corr['order_segment'] == segment]
    if len(subset) < 100:
        continue
    
    display(f"Сегмент: {segment} (n={len(subset)})")
    
    subset_features = [c for c in features if c not in ['total_orders', 'order_segment']]
    subset_data = subset[subset_features].dropna(subset=['avg_revenue_rub', 'avg_tickets', 'avg_days_between'])
    
    if len(subset_data) < 50:
        display("Мало данных")
        continue
    
    # Считаем корреляции внутри сегмента
    interval_in_subset = [c for c in interval_cols if c in subset_features]
    corr_seg = phik_matrix(subset_data, interval_cols=interval_in_subset)
    
    # Показываем топ-3 связи
    display("Топ-3 связи между признаками:")
    corr_list = []
    for i in range(len(corr_seg.columns)):
        for j in range(i+1, len(corr_seg.columns)):
            f1 = corr_seg.columns[i]
            f2 = corr_seg.columns[j]
            val = corr_seg.iloc[i, j]
            corr_list.append((f1, f2, val))
    
    # Сортируем по абсолютному значению
    corr_list_sorted = sorted(corr_list, key=lambda x: abs(x[2]), reverse=True)
    for f1, f2, val in corr_list_sorted[:3]:
        display(f"{f1} <-> {f2}: {val:.2f}")
````
